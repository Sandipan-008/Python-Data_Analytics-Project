# EXPLORATORY DATA ANALYSIS

### Countplot Chart 

```python
## Gender
ax = sns.countplot(x = 'Gender',data = df)
for bar in ax.containers:
    ax.bar_label(bar)
```

From above graphs we can see that most of the buyers are female and the purchasing power of women is greater than men

```python
## Age_Group
cz = sns.countplot(x = 'Age Group', data = df, hue = 'Gender')
for bars in cz.containers:
    cz.bar_label(bars)
```

From the above graph we can see that most of the buyers are of age group of between  26-35 females

```python
## Marital_Status
sd = sns.countplot(data = df, x = 'Marital_Status')
sns.set(rc= {'figure.figsize': (5,5)})
for bars in sd.containers:
    sd.bar_label(bars)
```

```python
## Occupation
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(x = 'Occupation',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

From the above graph we can see that most of the buyers are working in IT, HEALTHCARE, AVIATION sector

```python
## Product Category
sns.set(rc={'figure.figsize': (25,6)})
az = sns.countplot(x = 'Product_Category', data = df)
for bars in az.containers:
    az.bar_label(bars)
```

From the above graph we can see that most of the sold products are from Food, Clothing&Apparel, Electronics, footears

### Barplot Chart

```python
## Gender
sales_gen = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by = 'Amount',ascending=False)
bz = sns.barplot(x = 'Gender', y = 'Amount', data = sales_gen)

for bar in bz.containers:
    bz.bar_label(bar)
```

From above graphs we can see that most of the buyers are female and the purchasing power of women is greater than men

```python
## Age_Group
sales_age = df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by ='Amount', ascending = False)
sns.barplot(y = 'Amount',x = 'Age Group', data = sales_age)
```

From the above graph we can see that most of the buyers are of age group of between 26-35 females

```python
## Total Orders
sales_state = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize': (19, 8)})
sns.barplot(x = 'State', y = 'Orders', data = sales_state)
```

```python
## Total Sales
sales_state = df.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize': (19, 8)})
sns.barplot(x = 'State', y = 'Amount', data = sales_state)
```

From the above graph we can see that the most of the Orders & Sales are from UP, Maharashtra, & karnataka respectively

```python
## Total amount as per Marital_Status
sales_gen = df.groupby(['Gender','Marital_Status'],as_index=False)['Amount'].sum().sort_values(by = 'Amount',ascending=False)
sns.barplot(x = 'Marital_Status', y = 'Amount',hue = 'Gender', data = sales_gen)

```

From the above graph we can see that the most of the buyers are married[woman] & they have
the high purchasing power

```python
## Occupation
sales_state = df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=sales_state,x='Occupation', y='Amount')
```

From the above graph we can see that most of the buyers are working in IT, HEALTHCARE, AVIATION sector

```python
## Product Category
sales_category = df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(23,9)})
sns.barplot(x = 'Product_Category', y= 'Amount', data= sales_category)
```

From the above graph we can see that most of the sold products are from Food, Clothing&Apparel, Electronics, footears

```python
## Product_ID
sales_id = df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_id, x= 'Product_ID', y= 'Orders')
```
